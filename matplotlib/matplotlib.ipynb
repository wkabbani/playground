## Functional Plot

```python
x_1 = np.linspace(0, 5, 10)
y_1 = x_1**2
plt.plot(x_1, y_1)
plt.title('Plot 1')
plt.xlabel('time')
plt.ylabel('returns')
plt.show() # not needed in jupyter
```

## Multiple Plots

```python
plt.subplot(1, 2, 2)
plt.plot(x_1, y_1, 'r')
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, 'b')
```

## Figure Objects

```python
fig1 = plt.figure(figsize=(5,4))
axes1 = fig1.add_axes([0.1,0.1,0.9,0.9])
axes1.set_xlabel('axes 1 x label')
axes1.set_ylabel('axes 1 y label')
axes1.set_title('axes 1 title')
axes1.plot(x_1, y_1, label='plot1')
axes1.plot(y_1, x_1, label='plot2')
axes1.legend(loc=0)

axes2 = fig1.add_axes([0.45,0.45,0.4,0.3])
axes2.set_xlabel('axes 2 x label')
axes2.set_ylabel('axes 2 y label')
axes2.set_title('axes 2 title')
axes2.plot(x_1, y_1, label='plot1')
axes2.plot(y_1, x_1, label='plot2')
axes2.legend(loc=0)

axes2.text(10,35, 'text')
```

## SubPlots

```python
fig2, axes2 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
plt.tight_layout()
axes2[1].set_title('plot 2')
axes2[1].set_xlabel('x')
axes2[1].set_ylabel('y')
axes2[1].plot(x_1, y_1, 'r')

axes2[2].set_title('plot 3')
axes2[2].set_xlabel('x')
axes2[2].set_ylabel('y')
axes2[2].plot(x_1, y_1, 'b')

axes2[0].set_title('plot 1')
axes2[0].set_xlabel('x')
axes2[0].set_ylabel('y')
axes2[0].plot(x_1, y_1, 'y')
```

## Pie Charts

```python
import random

fig6 = plt.figure(figsize=(8,5))
axes6 = fig6.add_axes([0.1,0.1,0.9,0.9])

types = ['Water', 'Normal', 'Flying', 'Grass', 'Bug', 'Fire']
poke_num = [133, 109, 98, 30, 10, 50]

colors = []
for i in range(len(types)):
    rgb = (random.uniform(0,.5), random.uniform(0,.5), random.uniform(0,.5))
    colors.append(rgb)

explode = [0] * len(types)
explode[0] = 0.2

wedges, texts, autotexts = plt.pie(poke_num, explode=explode, labels=types, 
    colors=colors, autopct='%1.0f%%', shadow=True,
    startangle=140, textprops=dict(color='w'))

plt.legend(wedges, types, loc=0, bbox_to_anchor=(1, 0, 0.5, 1))
```

## 3D Surface

```python
from mpl_toolkits import mplot3d

fig9 = plt.figure(figsize=(8,5))
axes9 = fig9.add_axes([0.1,0.1,0.9,0.9], projection='3d')

z3 = 15 * np.random.randn(100)
x3 = np.sin(z3) * np.random.randn(100)
y3 = np.sin(z3) * np.random.randn(100)
axes9.scatter3D(x3, y3, z3, c=z3, cmap='Blues')
```

```python
def get_z(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x4 = np.linspace(-6, 6, 30)
y4 = np.linspace(-6, 6, 30)

x4, y4 = np.meshgrid(x4, y4)
z4 = get_z(x4, y4)

fig10 = plt.figure(figsize=(8,5))
axes10 = fig10.add_axes([0.1,0.1,0.9,0.9], projection='3d')
axes10.view_init(45, 55)
axes10.contour3D(x4, y4, z4, 100, cmap='Blues')
```

```python
fig11 = plt.figure(figsize=(8,5))
axes11 = fig11.add_axes([0.1,0.1,0.9,0.9], projection='3d')

axes11.set_xlabel('x')
axes11.set_xlabel('y')
axes11.set_xlabel('z')

axes11.plot_wireframe(x4, y4, z4, color='blue')
```

```python
fig12 = plt.figure(figsize=(8,5))
axes12 = fig12.add_axes([0.1,0.1,0.9,0.9], projection='3d')

axes12.set_xlabel('x')
axes12.set_xlabel('y')
axes12.set_xlabel('z')

axes12.plot_surface(x4, y4, z4, rstride=1, cstride=1, cmap='Blues', edgecolor=None)
```

```python
dimensions = ['one', 'two', 'three', 'four']
dates = ['June', 'July', 'August', 'Sep', 'Oct', 'Nov', 'Dec']
prts = np.array([
        [5.3, 5.5, 7.4, 6.1, 1.2, 2.3, 9.1],
        [3.3, 6.5, 7.4, 6.1, 1.2, 2.3, 8.1],
        [1.3, 5.5, 7.4, 0.1, 1.2, 7.3, 9.1],
        [8.3, 3.5, 8.4, 6.1, 1.2, 2.3, 3.1]
    ])
```

```python
fig13, axes13 = plt.subplots()
im = axes13.imshow(prts, cmap='Wistia')
axes13.set_xticks(np.arange(len(dates)))
axes13.set_yticks(np.arange(len(dimensions)))
axes13.set_xticklabels(dates)
axes13.set_yticklabels(dimensions)

plt.setp(axes13.get_xticklabels(), rotation=45, ha='right', rotation_mode='anchor')

for i in range(len(dimensions)):
    for j in range(len(dates)):
        text = axes13.text(j, i, prts[i,j], 
        ha='center', va='center', color='k',
        fontweight='bold')
```
