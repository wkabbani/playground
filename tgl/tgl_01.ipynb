## Images as Tensors

```python
import requests
from PIL import Image
from io import BytesIO

response = requests.get('https://images.pexels.com/photos/156934/pexels-photo-156934.jpeg?auto=compress&cs=tinysrgb&w=1260&h=750&dpr=1')
img = Image.open(BytesIO(response.content))
```

```python
img = Image.open(BytesIO(response.content)).convert('RGB').resize((256, 256))
img
```

```python
img_array = np.array(img)
```

```python
img_array.shape
```

```python
img_tensor = torch.tensor(img_array)
img_tensor.shape
```

```python
img_tensor = img_tensor.permute(2, 0, 1) / 255.0
img_tensor.shape
```

```python
img_tensor.min(), img_tensor.max()
```

```python
# or the same with torchvision
to_tensor_transform = torchvision.transforms.ToTensor()
img_tensor_1 = to_tensor_transform(img)
img_tensor_1.shape
```

```python
img_tensor_1.min(), img_tensor_1.max()
```

```python
img_tensor_1.unsqueeze(0).shape
```

```python
fix, axs = plt.subplots(1, 4, figsize=(16, 5))
axs[0].imshow(img_tensor.permute(1, 2, 0))
axs[0].set_title('original image')
colors = ['red', 'green', 'blue']
for i in range(1, 4):
  axs[i].imshow(img_tensor[i-1], cmap='gray')
  axs[i].set_title(f'{colors[i-1]} channel')
```
